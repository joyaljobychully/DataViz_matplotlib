1.   Plot y = x^2 titled 'X-Y' using normal method with x-axis name 'X' and y-axis name 'Y'

```python
x = np.arange(1,6)
```

```python
x
```

```python
y = x **2
```

```python
plt.plot(x,y)
plt.title('X-Y')
plt.xlabel('X')
plt.ylabel('Y')
```

2.   Make 3 plots of y=x^2 in the same figure with different colors


```python
plt.subplot(2,2,1)
plt.plot(x,y,'r')
plt.subplot(2,2,2)
plt.plot(x,y,'b')
plt.subplot(2,2,4)
plt.plot(x,y,'g')
```

3.   Plot y = x^2 using object oriented method. Name the axes and the graph

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y,'g*--')
axes.set_title('X-Y')
axes.set_xlabel('X')
axes.set_ylabel('Y')
```

4.   Make a plot of y=x**2 of line width 6 with 0.5 transparency and yellow point marks  of size 10 having black edges and edge size 5. Display only xE[3,13] and yE[10,120]. Inside the first plot there should be another plot with inverted axis  with scale and everything.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
##This is OOP in matlab
fig = plt.figure()
axis1 = fig.add_axes([0.1,0.1,0.8,0.8])    
axis2 = fig.add_axes([0.17,0.45,0.4,0.3])
x = np.arange(0,20)
y = x**2
axis1.plot(x,y,lw=6,ls='--',alpha=0.5,marker='o',markersize=10,markerfacecolor='yellow',
           markeredgewidth=5,markeredgecolor='black')
axis1.set_xlim(3,13)
axis2.set_ylim(10,120)
axis2.plot(y,x)
axis1.set_title('LARGER PLOT')
axis2.set_title('SMALLER PLOT')
axis1.set_xlabel('X')
axis1.set_ylabel('Y')
axis2.set_xlabel('Y')
axis2.set_ylabel('x')
```

5.   Make a subplot of 3 plots with x-x^2 in first axis and x^3-x in the ninth axis without overlap using object oriented method

```python
fig2,ax2 = plt.subplots(3,3)
ax2[0,0].plot(x,x**2)
ax2[2,2].plot(x**2,x)
plt.tight_layout()
```

6.   Make a plot of x-x^2 with figure size (8,2) and 150 dpi using object oriented method


```python
fig = plt.figure(figsize=(8,2),dpi=150)
ax1 = fig.add_axes([0,0,1,1])
ax1.plot(x,x**2,'b')
```

7.   Make a subplot of x-x^2 and x**3-x with figure size (8,2) and 150 dpi using object oriented method

```python
fig,ax1 = plt.subplots(3,3,figsize=(8,2),dpi=150)
ax1[0,0].plot(x,x**2)
ax1[2,2].plot(x**3,x)
plt.tight_layout()
```

8.   Make a canvas with both x-x^2 and x-x^3 in it. It should have a lengend to distinguish between the both.

```python
fig3 = plt.figure()
ax3 = fig3.add_axes([0,0,1,1])
ax3.plot(x,x**2,label='X SQUARED')
ax3.plot(x,x**3,label='X CUBED')
ax3.legend(loc=0)
```
