?2) Initialize two arrays- x with values from 0 to 10 and y with values from 11 to 20.


    1. Create a red dotted scatter plot with green title 'X-Y Scatter plot' in font size 20 , x axis name as 'X axis', y axis name as 'Y axis'. 
    2. Save the plot as jpeg file. Save the plot as a png file of 150 dpi.



```python
x = np.arange(10)
y = np.arange(11,21)
```

```python
plt.scatter(x,y,c='r')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('X-Y Scatter plot',fontsize = 20,c='g')
plt.savefig('xy_scatter.jpeg',dpi=150)
plt.show() ###Need to be put in actual code.
```

?6) Create a box plot for data = [np.random.normal(0, std, 100) for std in range(1, 4)]


    1. Create horizontal box plots
    2. Create colorless boxes


```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# rectangular box plot
plt.boxplot(data,vert=True,patch_artist=True)
```

```python
plt.boxplot(data,vert=False,patch_artist=True)
```

```python
plt.boxplot(data,vert=True,patch_artist=False)
```

?7) Plot the above pie chart.

```python
names = 'Python','Java','Ruby','C++'
separation = [0.4,0,0.2,0]
rang = ['yellow','skyblue','orange','red']
vals = [26.9,26.2,30.6,16.2]
plt.pie(vals,explode=separation,colors=rang,labels=names,shadow=True,autopct='%1.1f%%')
plt.axis('equal')
```
